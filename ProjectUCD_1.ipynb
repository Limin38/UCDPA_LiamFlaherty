```python
import pandas as pd
```

```python
contacts = pd.read_csv('./Contacts.csv')
Payments = pd.read_csv('./Payments.csv')
```

```python
contacts.info()
Payments.info()
```

```python
Contacts_final = contacts.merge(Payments, left_on = 'OrID', right_on = 'OrID')
```

```python
Contacts_final.info()
```

```python
Contacts_final.drop(['PaymentMethod_y'], axis= 'columns', inplace=True)

```

```python
Contacts_final.head()
```

```python
Contacts_final.drop_duplicates(subset="Order_ID",keep=False, inplace=True)
Contacts_final
```

```python
Contacts_final.sort_values("PaymentMethod", inplace = True)
```
